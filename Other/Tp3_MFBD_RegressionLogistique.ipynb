```r
# Chargement des données
#evans <- read.table("http://web1.sph.emory.edu/dkleinb/allDatasets/datasets/evans.dat")
evans <- read.csv("/content/evans.csv", header = TRUE)

# Inspection des données
head(evans)
str(evans)
summary(evans)

# # Vérification des valeurs manquantes
# sum(is.na(evans))
```

### 1.Analyse

```r
# Tableau de contingence pour la variable d'exposition principale
table(evans$CAT, evans$CDH)

# Statistiques descriptives de base
cat("Nombre total d'observations:", nrow(evans), "\n")
cat("Proportion de maladies coronariennes (chd):", mean(evans$CDH), "\n")
cat("Proportion avec catécholamine élevée (cat):", mean(evans$CAT), "\n")
```

### 2.Modele

```r
# Modèle logistique
modele <- glm(CDH ~ CAT + AGE + CHL + SMK + ECG + HPT,
              data = evans,
              family = binomial(link = "logit"))

summary(modele)
```

```r
exp(confint(modele))
```

### Interpreation et Resultats

```r
# Extraction des coefficients et calcul des Odds Ratios
coefficients <- coef(modele)
odds_ratios <- exp(coefficients)

# Création d'un tableau de résultats
resultats <- data.frame(
  Variable = names(coefficients),
  Coefficient = round(coefficients, 4),
  Odds_Ratio = round(odds_ratios, 4),
  p_value = round(summary(modele)$coefficients[,4], 4)
)

print(resultats)

# Interprétation spécifique pour la variable cat
cat("\n--- INTERPRÉTATION DE LA VARIABLE CAT ---\n")
cat("Odds Ratio pour cat (catécholamine):", round(odds_ratios["CAT"], 4), "\n")
cat("Cela signifie que les personnes avec un niveau de catécholamine élevé ont",
    round(odds_ratios["CAT"], 4), "fois plus de risque de maladie coronarienne,\n")
cat("après ajustement sur l'âge, le cholestérol, le tabagisme, l'ECG et l'hypertension.\n")
```

```r
# Test de déviance pour la significativité globale
cat("Déviance du modèle:", deviance(modele), "\n")
cat("Déviance du modèle nul:", modele$null.deviance, "\n")

# Pseudo R²
pseudo_r2 <- 1 - (deviance(modele) / modele$null.deviance)
cat("Pseudo R²:", round(pseudo_r2, 4), "\n")

# Test du rapport de vraisemblance
anova(modele, test = "Chisq")
```

```r
cat("\n=== RÉSULTATS PRINCIPAUX ===\n")
cat("Variable d'exposition: cat (niveau de catécholamine)\n")
cat("Variables de contrôle: age, chl, smk, ecg, hpt\n")
cat("Effectif total:", nrow(evans), "hommes\n\n")

cat("Résultats du modèle logistique:\n")
print(resultats)

cat("\nLe modèle montre que la variable 'cat' a un coefficient de",
    round(coefficients["CAT"], 4), "et un Odds Ratio de",
    round(odds_ratios["CAT"], 4), "\n")
```
