```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

```

```python
model_class_bin=Sequential()
model_class_bin.add(Dense(4,input_shape=(4,),activation='tanh'))
model_class_bin.add(Dense(1,activation="sigmoid"))
```

```python
model_class_bin.summary()
```

```python
model_class_bin.compile(optimizer="sgd",loss="binary_crossentropy",metrics=[])
```

```python
import numpy as np
```

```python
X_train=np.load('/content/drive/MyDrive/X_train_banknotes.npy')
X_test=np.load('/content/drive/MyDrive/X_test_banknotes.npy')
y_train=np.load('/content/drive/MyDrive/y_train_banknotes.npy')
y_test=np.load('/content/drive/MyDrive/y_test_banknotes.npy')
```

```python
model_class_bin.fit(X_train,y_train,epochs=30,batch_size=16)
```

```python
model_class_bin.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

# Get predictions for the test set
y_pred_prob = model_class_bin.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype("int32")

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy: {accuracy}")
```
